```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import OrdinalEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score,classification_report
```

```python
iris=pd.read_csv('/content/drive/MyDrive/Iris.csv')
iris
```

```python
IrIs=iris
```

```python
iris.isnull().sum()
```

```python
encoder=OrdinalEncoder()
Iris=encoder.fit_transform(iris[['Species']])
Iris
```

```python
iris['Species']=Iris
iris
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called "IrIs" with columns "SepalWidthCm", "SepalLengthCm", and "Species"

# Count the occurrences of each category on the x-axis
sns.countplot(data=IrIs, x='SepalWidthCm')

# Set labels and title
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Count')
plt.title('Bar Plot of Sepal Width')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called "IrIs" with columns "SepalWidthCm", "SepalLengthCm", and "Species"

# Create a bar plot with the mean of Sepal Length for each category on the x-axis
sns.barplot(data=IrIs, x='SepalWidthCm', y='SepalLengthCm')

# Set labels and title
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Mean Sepal Length (cm)')
plt.title('Bar Plot of Sepal Length by Sepal Width')

# Display the plot
plt.show()
```

```python
sns.scatterplot(data=IrIs[:],x='PetalWidthCm',y='PetalLengthCm',hue="Species")
```

```python
k_means=KMeans(n_clusters=3)
k_means.fit(iris)
res=k_means.predict([[0,5.9,3.0,5.1,1.8,0]])
print(res)
kmeans_labels = k_means.labels_
kmeans_score = silhouette_score(iris, kmeans_labels)
print('\n\nK-Means Silhouette Score:', kmeans_score)
print()
print()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z = linkage(iris, method='centroid')
```

```python
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca_iris = pca.fit_transform(iris)
pca_iris
```

```python
plt.scatter(pca_iris[:, 0], pca_iris[:, 1], c=iris['Species'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Iris Dataset')
plt.show()
```
